```python
import csv
import os
import requests
import tqdm

def download_pdfs_from_csv(csv_file, directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

    with open(csv_file, 'r') as file:
        reader = csv.reader(file)
        for row in tqdm.tqdm(reader):
            try:
                pdf_url = row[20]  # Assuming the URL is in the first column of each row
                download_pdf(pdf_url, directory)
            except:
                pass

def download_pdf(url, directory):
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
}
    response = requests.get(url,headers=headers)
    if response.status_code == 200:
        # Extract the filename from the URL
        filename = os.path.basename(url)
        file_path = os.path.join(directory, filename)
        with open(file_path, 'wb') as file:
            file.write(response.content)
            print(f"Downloaded: {filename}")
    else:
        print(f"{response.status_code} Failed to download PDF from: {url}")

# Usage
csv_file_path = "C:/Users/punit/Downloads/Data Engineer (Quantitative Research Desk) - Internship + Full Time (Gurgaon C&E)_09052023_19052023.csv"
download_directory = "./DataEngineer"
download_pdfs_from_csv(csv_file_path, download_directory)
```

```python
import os
import re
import PyPDF2

def filter_pdfs_by_keyword(directory, keywords):
    filtered_files = []

    # Iterate over all files in the directory
    for filename in tqdm.tqdm(os.listdir(directory)):
        if filename.endswith('.pdf'):
            file_path = os.path.join(directory, filename)
            if has_keyword(file_path, keywords):
                filtered_files.append(filename)

    return filtered_files

def has_keyword(file_path, keywords):
    try:
        with open(file_path, 'rb') as file:
            pdf_reader = PyPDF2.PdfReader(file)
            num_pages = pdf_reader.pages
            for page_num in range(len(num_pages)):
                page = pdf_reader.pages[page_num]
                page_text = page.extract_text()

                for keyword in keywords:
                    pattern = r'\b{}\b'.format(re.escape(keyword))
                    if re.search(pattern, page_text, re.IGNORECASE):
                        return True
    except:
        pass

    return False

# Usage
pdf_directory = './DataEngineer'
search_keywords = ['Airflow']

filtered_files = filter_pdfs_by_keyword(pdf_directory, search_keywords)
print("Filtered PDF files:")
for file in filtered_files:
    print(file)
```

```python
import os
import pandas as pd
import fitz
import requests
import textract
import tempfile
import io
import re
import shutil
import tkinter as tk
from tkinter import filedialog
from tkinter import ttk
from PIL import ImageTk, Image

# Global variables for GUI elements
root = None
frame = None
excel_path_label = None
keywords_entry = None
save_path_entry = None
dataframe_table = None

def browse_excel_file():
    file_path = filedialog.askopenfilename(filetypes=[("Excel Files", "*.xlsx")])
    if file_path:
        excel_path_label.config(text=file_path)

def create_dataframe_table(dataframe):
    # Clear any existing table
    clear_dataframe_table()

    # Create a new table frame
    table_frame = ttk.Frame(frame)
    table_frame.pack(pady=10)

    # Create a treeview widget as the table
    treeview = ttk.Treeview(table_frame)
    treeview['columns'] = list(dataframe.columns)
    treeview['show'] = 'headings'

    # Add columns to the table
    for column in dataframe.columns:
        treeview.heading(column, text=column)
        treeview.column(column, width=100)

    # Add rows to the table
    for row in dataframe.itertuples(index=False):
        treeview.insert('', 'end', values=row)

    # Add the table to a scrollable frame
    scrollbar = ttk.Scrollbar(table_frame, orient="vertical", command=treeview.yview)
    treeview.configure(yscrollcommand=scrollbar.set)
    scrollbar.pack(side='right', fill='y')
    treeview.pack()

    # Update the global reference to the table frame
    global dataframe_table
    dataframe_table = table_frame

def clear_dataframe_table():
    global dataframe_table
    if dataframe_table:
        dataframe_table.destroy()
        dataframe_table = None

def download_resume(url, save_directory):
    try:
        filename = os.path.basename(url)
        file_path = os.path.join(save_directory, filename)
        if os.path.exists(file_path):
            print(f"Resume already exists: {filename}")
            return
        # Check if the URL is actually a local file path
        if os.path.exists(url):
            shutil.copy(url, file_path)
            print(f"Downloaded: {filename}")
            return
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.82 Safari/537.36'
            }
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded: {filename}")
        else:
            print(f"Failed to download resume from: {url}")
    except:
        pass

def convert_to_pdf(file_path):
    # Generate a temporary PDF file path
    temp_pdf_path = tempfile.mktemp(suffix='.pdf')

    try:
        # Convert non-PDF resumes to PDF using textract
        text = textract.process(file_path)
        with open(temp_pdf_path, 'wb') as temp_pdf:
            temp_pdf.write(text)
        return temp_pdf_path
    except Exception as e:
        print(f"Failed to convert file to PDF: {file_path}")
        return None

def filter_dataframe_by_keyword(dataframe, keywords):
    filtered_df = dataframe.copy()

    filtered_df = filtered_df.dropna(subset=['Resume Path'])

    filtered_df['Has Keyword'] = filtered_df['Resume Path'].apply(lambda x: has_keyword(x, keywords))
    filtered_df = filtered_df[filtered_df['Has Keyword']]

    return filtered_df

def has_keyword(file_path, keywords):
    try:
        with fitz.open(file_path) as doc:
            for page in doc:
                try:
                    page_text = page.get_text()
                    for keyword in keywords:
                        pattern = r'\b{}\b'.format(re.escape(keyword))
                        if re.search(pattern, page_text, re.IGNORECASE):
                            return True
                except Exception as e:
                    print(f"Error occurred while processing page in PDF file: {file_path}")
    except Exception as e:
        print(f"Error occurred while processing PDF file: {file_path}")
    
    return False

def browse_save_path():
    save_path = filedialog.askdirectory()
    if save_path:
        save_path_entry.delete(0, tk.END)
        save_path_entry.insert(0, save_path)

def process_excel_file():
    # Get the paths and keywords from the GUI inputs
    excel_file_path = excel_path_label['text']
    keywords = keywords_entry.get().split(',')
    save_path = save_path_entry.get()

    # Read the Excel file into a DataFrame
    df = pd.read_excel(excel_file_path)

    # Download resumes and store the file paths
    resume_directory = os.path.join(save_path, 'resumes')
    if not os.path.exists(resume_directory):
        os.makedirs(resume_directory)

    df['Resume Path'] = df['File Upload'].apply(lambda x: os.path.join(resume_directory, os.path.basename(x)))
    df['Resume Downloaded'] = df['File Upload'].apply(lambda x: download_resume(x, resume_directory))

    # Filter the DataFrame based on keywords
    filtered_df = filter_dataframe_by_keyword(df, keywords)

    # Display the filtered DataFrame as a table
    create_dataframe_table(filtered_df)

def main():
    global root, frame, excel_path_label, keywords_entry, save_path_entry

    # Create the main application window
    root = tk.Tk()
    root.title("Resume Filter")
    root.geometry("800x600")

    # Create a frame to hold the content
    frame = tk.Frame(root)
    frame.pack(pady=20)

    # Excel file selection
    excel_label = tk.Label(frame, text="Select Excel File:")
    excel_label.pack()

    excel_frame = tk.Frame(frame)
    excel_frame.pack()

    excel_path_label = tk.Label(excel_frame, text="")
    excel_path_label.pack(side=tk.LEFT, padx=(0, 10))

    excel_button = tk.Button(excel_frame, text="Browse", command=browse_excel_file)
    excel_button.pack()

    # Keywords entry
    keywords_label = tk.Label(frame, text="Enter Keywords (comma-separated):")
    keywords_label.pack()

    keywords_entry = tk.Entry(frame, width=50)
    keywords_entry.pack()

    # Save path selection
    save_path_label = tk.Label(frame, text="Select Save Path:")
    save_path_label.pack()

    save_path_frame = tk.Frame(frame)
    save_path_frame.pack()

    save_path_entry = tk.Entry(save_path_frame, width=50)
    save_path_entry.pack(side=tk.LEFT, padx=(0, 10))

    save_path_button = tk.Button(save_path_frame, text="Browse", command=browse_save_path)
    save_path_button.pack()

    # Process button
    process_button = tk.Button(frame, text="Process Excel File", command=process_excel_file)
    process_button.pack(pady=20)

    # Start the Tkinter event loop
    root.mainloop()

if __name__ == "__main__":
    main()
```
